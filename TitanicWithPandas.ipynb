```python
df['Gender'] = df["Sex"].map( { "female": 0, "male": 1 } )
```

```python
df['Gender'] = df["Sex"].map( { "female": 0, "male": 1 } ).astype(int)
```

```python
df.head()
```

```python
df['Gender2'] = df['Sex'].apply(lambda sex: "1" if sex == "male" else "0")
```

```python
df.head()
```

```python
def numeric_gender(sex):
    if sex == 'female':
        return 0
    if sex == 'male':
        return 1

df["Gender"] = df["Sex"].apply(numeric_gender)
df[["Sex", "Gender"]]
```

```python
df = df.drop(['Gender2'], axis=1)
```
